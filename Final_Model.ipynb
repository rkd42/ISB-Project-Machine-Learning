```python
X = da.iloc[:, 2:] #Input data
y = da.iloc[:, 0]  #Output data
```

```python
#Train and Test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.03, random_state = 0)
```

```python
#Data Standardization
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

# Creating functions for models to be used

```python
def catboost_model(iterations, learning_rate, loss_function,bagging_temperature,
                             border_count,l2_leaf_reg,max_depth,random_strength):
  
  try:
    from catboost import CatBoostClassifier
  except:
    !pip install catboost
    from catboost import CatBoostClassifier

  model = CatBoostClassifier(iterations=iterations, learning_rate=learning_rate, loss_function=loss_function,bagging_temperature=bagging_temperature,
                             border_count= border_count,l2_leaf_reg=l2_leaf_reg,max_depth=max_depth,random_strength=random_strength)

  model.fit(X_train, y_train,logging_level='Silent')


  acc_ = model.score(X_test,y_test)
  
  predi_ =  model.predict(X_test)

  print("catboost's prediction accuracy is: ", acc_)
  

  return model,predi_

def Randomforest_model(estimators):
  
  try:
    from sklearn.ensemble import RandomForestClassifier
  except:
    pass

  model = RandomForestClassifier(n_estimators=estimators, criterion='entropy', random_state=0,n_jobs = -1,min_samples_leaf = 1)
  model.fit(X_train, y_train)

  acc_ = model.score(X_test,y_test)
  
  predi_ =  model.predict(X_test)

  print("Randomforest's prediction accuracy is: ", acc_)
  
      
  return model,predi_

def XGB_model(colsample_bytree,nthread,subsample,n_estimators,eta,gamma,max_depth,learning_rate):
  
  try:
    import xgboost as xgb
  except:
    !pip install xgboost
    import xgboost as xgb

  model = xgb.XGBClassifier(colsample_bytree=colsample_bytree,learning_rate = learning_rate, 
                             max_depth = max_depth, n_estimators = n_estimators, 
                             nthread=nthread,eta=eta,gamma=gamma,subsample = subsample)
  model.fit(X_train, y_train)

  acc_ = model.score(X_test,y_test)
  
  predi_ =  model.predict(X_test)

  print("XGBOOST's prediction accuracy is: ", acc_)
  
  return model,predi_
```

# Fiting data on models

```python
#fiting on Catboost
catboost_model,catboost_model_prediction = catboost_model(iterations=80000, learning_rate=0.02, loss_function='CrossEntropy',bagging_temperature=0.4778,
                             border_count=238,l2_leaf_reg=3.347,max_depth=10,random_strength=0.01918)
```

```python
#fiting on Randomforest
Randomforest_model,Randomforest_model_prediction = Randomforest_model(estimators=400)
```

```python
#fiting on XGboost
XGB_model,XGB_model_prediction=XGB_model(colsample_bytree=0.9,nthread=2,
                        subsample=0.9,n_estimators=800,eta=0.1,gamma=0.0,max_depth=76,learning_rate=0.2)
```

# Converting predicted values to dataframes

```python
#catboost's predition
dataframe_cat=pd.DataFrame(catboost_model_prediction, columns=['catboost_model_prediction']) 
#Randomforest's prediction
dataframe_ran=pd.DataFrame(Randomforest_model_prediction , columns=['Randomforest_model_prediction '])
#XGboost's prediction
dataframe_xg=pd.DataFrame(XGB_model_prediction, columns=['XGB_model_prediction']) 
#Merging the three into a single dataframe
dataframe=pd.concat([dataframe_cat,dataframe_ran,dataframe_xg], axis=1)



```

# Ensembling Explained

We first add the predictions of all three models,and use majority vote criteria to assign the final predicted value 
i.e. if the (sum of outcomes) <=1 we take the predicted value as 0 and if the (sum of outcomes) >=2 we take it as 1

```python
dataframe_sum=dataframe.sum(axis = 1) #creating a dataframe with sum of all predicted values
dataframe=dataframe_sum.values 
dataframe_final=pd.DataFrame(dataframe, columns=['output'])
dataframe_final.loc[dataframe_final.output == 1, 'output'] = 0 #converting 1's to 0's
dataframe_final.loc[dataframe_final.output == 2, 'output'] = 3 #converting 2's to 3's
dataframe_final.loc[dataframe_final.output == 3, 'output'] = 1 #converting 3's to 1's
```

#  final accuracy

```python
accuracy = accuracy_score(y_test,dataframe_final) 
print('Model_AUC_score', accuracy)
```
