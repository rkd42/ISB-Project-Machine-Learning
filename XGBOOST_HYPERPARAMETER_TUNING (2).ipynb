```python
X = da.iloc[:, 2:] #Input data
y = da.iloc[:, 0] #Output data
```

```python
#Train and Test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.03, random_state = 0)
```

```python
#Data Standardization
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

# Hyperparameter_optimization

```python
#bounds on different parameters 
param_to_be_optimized = {'eta':(0.1,0.5),'gamma':(0,3),'max_depth':(10,100),'learning_rate':(0.1,0.8),
                        'n_estimators':(50,200),'subsample':(0.1,0.9),'colsample_bytree':(0.1,0.9),'nthread':(1,10)}





def param_handler_to_optimize(eta,gamma,max_depth,learning_rate,n_estimators,subsample,colsample_bytree,nthread):
    """
    To handle integer type parameters:
    
    """
    max_depth = int(max_depth) #int type params
    n_estimators = int(n_estimators)# int type params
    nthread = int(nthread)#int type params
    
    param = {
    'max_depth': max_depth,  # the maximum depth of each tree
    'eta': eta,  # the training step for each iteration
    'silent': 1,  # logging mode - quiet
    'objective': 'binary:logistic',  # error evaluation for multiclass training
    
    'n_estimators':10,
    'gamma':gamma,
    'learning_rate':learning_rate,
    'n_estimators':n_estimators,
    'subsample':subsample,
    'colsample_bytree':colsample_bytree,
    'nthread':nthread}
    return func_to_be_optimized(param)

def func_to_be_optimized(param):
    
    model = xgb.XGBClassifier(**param)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return accuracy_score(y_test,preds) 

optimizer = BayesianOptimization(
    f=param_handler_to_optimize,
    pbounds=param_to_be_optimized,
    random_state=1,
)
optimizer.maximize(
    init_points=2,
    n_iter=100,
)
```
